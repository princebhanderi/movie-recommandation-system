```python
import numpy as np 
import pandas as pd
```

```python
movie=pd.read_csv('tmdb_5000_movies.csv')
credit=pd.read_csv('tmdb_5000_credits.csv')
```

```python
df=movie.merge(credit,on='title')
```

```python
df.head()
```

```python
df.columns
```

```python
df=df[['movie_id','genres','keywords','title','overview','cast','crew']]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
import ast
```

```python
def convert(x):
    l=[]
    for i in ast.literal_eval(x):
        l.append(i['name'])
    return l
```

```python
df['genres']=df['genres'].apply(convert)
df['keywords']=df['keywords'].apply(convert)
df['cast']=df['cast'].apply(convert)
```

```python
df['cast'][1]
```

```python
df['cast']=df['cast'].apply(lambda x:x[0:3])
```

```python
df['cast'][0]
```

```python
def director(x):
    l=[]
    for i in ast.literal_eval(x):
        if(i['job']=='Director'):
            l.append(i['name'])
    return l
```

```python
df['crew']=df['crew'].apply(director)
```

```python
df['crew'][0]
```

```python
df.head()
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df['overview']
```

```python
def collapse(x):
    l=[]
    for i in x:
        l.append(i.replace(" ",""))
    return l
```

```python
df['genres']=df['genres'].apply(collapse)
df['keywords']=df['keywords'].apply(collapse)
df['cast']=df['cast'].apply(collapse)
df['crew']=df['crew'].apply(collapse)
```

```python
df['tag']=df['genres']+df['keywords']+df['overview']+df['cast']+df['crew']
```

```python
df['tag']=df['tag'].apply(lambda x:" ".join(x))
```

```python
df['tag'][0]
```

```python
df['cast'][0]
```

```python
new=df[['movie_id','title','tag']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(name):
    index=new[new['title']==name].index[0]
    movies=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in movies[0:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
